## Real-Time Face Detection from Webcam

The below code will not run in google colab because here the local camera is not accessable.

```python
import cv2
import face_recognition

# Capture the video from default camera
webcam_video_stream = cv2.VideoCapture(0)

# Initialize the array variable to hold all face locations in the frame
all_face_locations = []

# Looping through every frame in the video
while True:
    # Get  the current frame from thevideo stream as an image
    ret, current_frame = webcam_video_stream.read()
    # resize the current frame to 1/4 size to process faster
    current_frame_small= cv2.resize(current_frame,(0,0),fx=0.25,fy=0.25)

    # Detect all faces in the image
    # Arguments are image, no_of_times_to_upsample, model
    # all_face_locations = face_recognition.face_locations(current_frame_small, model='hog')
    # Without number_of_times_to_upsample(by default 1), it will not detect face from far.
    all_face_locations = face_recognition.face_locations(current_frame_small, number_of_times_to_upsample=2, model='hog')

    # Looping through the face locations
    for index, current_face_location in enumerate(all_face_locations):
        # Splitting the tuple to get the four position values
        top_pos, right_pos, bottom_pos, left_pos = current_face_location

        # Change the position magnitude to fit the actual size video frame
        top_pos = top_pos*4
        right_pos = right_pos*4
        bottom_pos = bottom_pos*4
        left_pos = left_pos*4

        # Printing the location of current face
        print('Found Face {} at top:{}, right:{}, bottom:{}, left:{}'.format(index+1, top_pos, right_pos, bottom_pos, left_pos))

        # Draw rectangle around the face detected
        cv2.rectangle(current_frame, (left_pos, top_pos), (right_pos, bottom_pos), (0, 0, 255), 2)

    # Showing the current face with rectangle drawn
    # cv2.imshow("Webcam Video", current_frame)
    cv2_imshow(current_frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the stream and cam
# Close all open openCV windows
webcam_video_stream.release()
cv2.destroyAllWindows()
```

## Real-Time Facial Expression Detection

```python
# Installing Libraries

# Using the Neural Network Library Backend TensorFlow
# Keras, an easy to use wrapper to the low-level library TensorFlow
# Before installing Keras, install the backend engine: TensorFlow

# https://www.kaggle.com/c/challenges-in-representation-learning-facial-expression-recognition-challenge/data?select=fer2013.tar.gz
```

```python
# Step-1 : Importing the required libraries
import numpy as np
import cv2
from keras.preprocessing import image
from keras.models import model_from_json
import face_recognition


# Capture the video from default camera
webcam_video_stream = cv2.VideoCapture(0)


# Step-2 : Initialize model and load weights

# Face expression model initialization
face_exp_model = model_from_json(open("facial_expression_model_structure.json", "r").read())

# Load weights into model
face_exp_model.load_weights('facial_expression_model_weights.h5')

# List of emotions labels
emotions_label = ('angry', 'disgust', 'fear', 'happy', 'sad', 'surprise', 'neutral')

# Initialize the array variable to hold all face locations in the frame
all_face_locations = []

# Looping through every frame in the video
while True:

    # Get  the current frame from thevideo stream as an image
    ret, current_frame = webcam_video_stream.read()

    # Resize the current frame to 1/4 size to process faster
    current_frame_small= cv2.resize(current_frame,(0,0),fx=0.25,fy=0.25)

    # Detect all faces in the image
    # Arguments are image, no_of_times_to_upsample, model
    # all_face_locations = face_recognition.face_locations(current_frame_small, model='hog')
    # Without number_of_times_to_upsample(by default 1), it will not detect face from far.
    all_face_locations = face_recognition.face_locations(current_frame_small, number_of_times_to_upsample=2, model='hog')

    # Looping through the face locations
    for index, current_face_location in enumerate(all_face_locations):

        # Splitting the tuple to get the four position values
        top_pos, right_pos, bottom_pos, left_pos = current_face_location

        # Change the position magnitude to fit the actual size video frame
        top_pos = top_pos*4
        right_pos = right_pos*4
        bottom_pos = bottom_pos*4
        left_pos = left_pos*4

        # Printing the location of current face
        print('Found Face {} at top:{}, right:{}, bottom:{}, left:{}'.format(index+1, top_pos, right_pos, bottom_pos, left_pos))


        # Step-3 : Slice and extract the face image from frame

        # Slicing the current face from main image
        current_face_image = current_frame[top_pos:bottom_pos, left_pos:right_pos]

        # Draw rectangle around the face detected
        cv2.rectangle(current_frame, (left_pos, top_pos), (right_pos, bottom_pos), (0, 0, 255), 2)


        # Step-4 : Pre-process the input image (video frame)

        # Preprocess input, convert it to an image like as the data in dataset
        # Convert to grayscale
        current_face_image = cv2.cvtColor(current_face_image, cv2.COLOR_BGR2GRAY)

        # Resize to 48x48 px size
        current_face_image = cv2.resize(current_face_image, (48, 48))

        # Convert the PIL image into a 3D numpy array
        img_pixels = image.img_to_array(current_face_image)

        # Expand the shape of an array into single row multiple columns
        img_pixels = np.expand_dims(img_pixels, axis = 0)

        # Pixels are in range of [0, 255]. Normalize all pixels in scale of [0, 1]
        img_pixels /= 255


        # Step-5 : Do prediction and display results

        # Do prediction using model, get the prediction values for all 7 expressions
        exp_predictions = face_exp_model.predict(img_pixels)

        # Find max indexed prediction value (0 till 7)
        max_index = np.argmax(exp_predictions[0])

        # Get corresponding lable from emotions_label
        emotion_label = emotions_label[max_index]

        # Display the name as text in the image
        font = cv2.FONT_HERSHEY_DUPLEX
        cv2.putText(current_frame, emotion_label, (left_pos, bottom_pos), font, 0.5, (255, 255, 255), 1)

    # Showing the current face with rectangle drawn
    # cv2.imshow("Webcam Video", current_frame)
    cv2_imshow(current_frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the stream and cam
# Close all open openCV windows
webcam_video_stream.release()
cv2.destroyAllWindows()
```

## Real-Time Age and Gender Classification

Downloading the Pre-Trained Model

Model Created and Trained by Gil Levi and Tal Hassner using the Adience dataset.
https://talhassner.github.io/home/projects/Adience/Adience-data.html

The trained Model files can be downloaded from the source :\
https://talhassner.github.io/home/publication/2015_CVPR

```python
import cv2
import face_recognition

# Capture the video from default camera
webcam_video_stream = cv2.VideoCapture(0)

# Initialize the array variable to hold all face locations in the frame
all_face_locations = []

# Looping through every frame in the video
while True:
    # Get  the current frame from thevideo stream as an image
    ret, current_frame = webcam_video_stream.read()
    # resize the current frame to 1/4 size to process faster
    current_frame_small= cv2.resize(current_frame,(0,0),fx=0.25,fy=0.25)

    # Detect all faces in the image
    # Arguments are image, no_of_times_to_upsample, model
    # all_face_locations = face_recognition.face_locations(current_frame_small, model='hog')
    # Without number_of_times_to_upsample(by default 1), it will not detect face from far.
    all_face_locations = face_recognition.face_locations(current_frame_small, number_of_times_to_upsample=2, model='hog')

    # Looping through the face locations
    for index, current_face_location in enumerate(all_face_locations):
        # Splitting the tuple to get the four position values
        top_pos, right_pos, bottom_pos, left_pos = current_face_location

        # Change the position magnitude to fit the actual size video frame
        top_pos = top_pos*4
        right_pos = right_pos*4
        bottom_pos = bottom_pos*4
        left_pos = left_pos*4

        # Printing the location of current face
        # print('Found Face {} at top:{}, right:{}, bottom:{}, left:{}'.format(index+1, top_pos, right_pos, bottom_pos, left_pos))
        
        # Slicing the current face from main image
        current_face_image = current_frame[top_pos:bottom_pos, left_pos:right_pos]

        
        # Step-1 : Declare mean value
        
        # The 'AGE_GENDER_MODEL_MEAN_VALUES' calculated by using the numpy.mean()
        AGE_GENDER_MODEL_MEAN_VALUES = (78.4263377603, 87.7689143744, 114.895847746)

        
        # Step-2 : Create blob of face slice

        # Create blob of current face slice
        # Params (image, scale, (size), (mean), RBSwap)
        current_face_image_blob = cv2.dnn.blobFromImage(current_face_image, 1, (227, 227), AGE_GENDER_MODEL_MEAN_VALUES, swapRB=False)

        
        # Step-3 : Declare gender labels, protext and caffemodel file paths
        gender_label_list = ['Male', 'Female']
        gender_protext = "gender_deploy.prototxt"
        gender_caffemodel = "gender_net.caffemodel"


        # Step-4 : Create model from files and provide blob as input
        gender_cov_net = cv2.dnn.readNet(gender_caffemodel, gender_protext)
        gender_cov_net.setInput(current_face_image_blob)


        # Step-5 : Get gender predictions and get label of maximum value returned item
        gender_predictions = gender_cov_net.forward()
        gender = gender_label_list[gender_predictions[0].argmax()]

        
        # Step-6 : Repeat Same Steps for Age Prediction
        age_label_list = ['(0-2)', '(4-6)', '(8-12)', '(15-20)', '(25-32)', '(38-43)', '(48-53)', '(60-100)']
        age_protext = "age_deploy.prototxt"
        age_caffemodel = "age_net.caffemodel"
        age_cov_net = cv2.dnn.readNet(age_caffemodel, age_protext)
        age_cov_net.setInput(current_face_image_blob)
        age_predictions = age_cov_net.forward()
        age = age_label_list[age_predictions[0].argmax()]


        # Step-7 : Print gender and age under the rectangle

        # Draw rectangle around the face detected
        cv2.rectangle(current_frame, (left_pos, top_pos), (right_pos, bottom_pos), (0, 0, 255), 2)

        # Display the name as text in the image
        font = cv2.FONT_HERSHEY_DUPLEX
        cv2.putText(current_frame, gender+" "+age+" "+"yrs", (left_pos, bottom_pos+20), font, 0.5, (0, 255, 0), 1)

    # Showing the current face with rectangle drawn
    # cv2.imshow("Webcam Video", current_frame)
    cv2_imshow(current_frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the stream and cam
# Close all open openCV windows
webcam_video_stream.release()
cv2.destroyAllWindows()
```

## Image Age and Gender Detection

```python
# Step-1 : Importing the required libraries
import cv2
import face_recognition

# Loading the image to detect
image_to_detect = cv2.imread('trump-modi.jpg')

# Detect all faces in the image
# Arguments are image, no_of_times_to_upsample, model
all_face_locations = face_recognition.face_locations(image_to_detect, model='hog')

# Print the number of faces detected
print("There are {} face(s) in this image.".format(len(all_face_locations)))

# Looping through the face locations
for index, current_face_location in enumerate(all_face_locations):
    # Splitting the tuple to get the four position values
    top_pos, right_pos, bottom_pos, left_pos = current_face_location
    
    # Printing the location of current face
    print('Found Face {} at top:{}, right:{}, bottom:{}, left:{}'.format(index+1, top_pos, right_pos, bottom_pos, left_pos))

    # Slicing the current face from main image
    current_face_image = image_to_detect[top_pos:bottom_pos, left_pos:right_pos]

    # Step-1 : Declare mean value
        
    # The 'AGE_GENDER_MODEL_MEAN_VALUES' calculated by using the numpy.mean()
    AGE_GENDER_MODEL_MEAN_VALUES = (78.4263377603, 87.7689143744, 114.895847746)

    
    # Step-2 : Create blob of face slice

    # Create blob of current face slice
    # Params (image, scale, (size), (mean), RBSwap)
    current_face_image_blob = cv2.dnn.blobFromImage(current_face_image, 1, (227, 227), AGE_GENDER_MODEL_MEAN_VALUES, swapRB=False)

    
    # Step-3 : Declare gender labels, protext and caffemodel file paths
    gender_label_list = ['Male', 'Female']
    gender_protext = "gender_deploy.prototxt"
    gender_caffemodel = "gender_net.caffemodel"


    # Step-4 : Create model from files and provide blob as input
    gender_cov_net = cv2.dnn.readNet(gender_caffemodel, gender_protext)
    gender_cov_net.setInput(current_face_image_blob)


    # Step-5 : Get gender predictions and get label of maximum value returned item
    gender_predictions = gender_cov_net.forward()
    gender = gender_label_list[gender_predictions[0].argmax()]

    
    # Step-6 : Repeat Same Steps for Age Prediction
    age_label_list = ['(0-2)', '(4-6)', '(8-12)', '(15-20)', '(25-32)', '(38-43)', '(48-53)', '(60-100)']
    age_protext = "age_deploy.prototxt"
    age_caffemodel = "age_net.caffemodel"
    age_cov_net = cv2.dnn.readNet(age_caffemodel, age_protext)
    age_cov_net.setInput(current_face_image_blob)
    age_predictions = age_cov_net.forward()
    age = age_label_list[age_predictions[0].argmax()]


    # Step-7 : Print gender and age under the rectangle

    # Draw rectangle around the face detected
    cv2.rectangle(image_to_detect, (left_pos, top_pos), (right_pos, bottom_pos), (0, 0, 255), 2)

    # Display the name as text in the image
    font = cv2.FONT_HERSHEY_DUPLEX
    cv2.putText(image_to_detect, gender+" "+age+" "+"yrs", (left_pos, bottom_pos+20), font, 0.5, (0, 255, 0), 1)


    # cv2.imshow("Gender and Age on Image ", image_to_detect)
    cv2_imshow(image_to_detect)
```

But Face Recognition relies on 128 points or 128 numbers called as Face Embeddings and these face embeddings is used to compare our face with what we already have.

## Image Face Recognition

```python
# Re-using the Image Face Detection Code

import cv2
import face_recognition


# Loading the image to detect
original_image = cv2.imread('trump-modi.jpg')


# Step-1 : Load Sample Images and Extract the Face Encodings

# Load a sample picture and and extract face encodings.
# Return a list of 128-dimensional face encodings (one for eace face in the image).
# We are getting the first face (assuming only one face).
modi_image = face_recognition.load_image_file("Samples/modi.jpg")
modi_face_encoding = face_recognition.face_encodings(modi_image)[0]

trump_image = face_recognition.load_image_file("Samples/trump.jpg")
trump_face_encoding = face_recognition.face_encodings(trump_image)[0]


# Step-2 : Create an array to save the encodings

known_face_encodings = [modi_face_encoding, trump_face_encoding]


# Step-3 : Create another array to hold labels

known_face_names = ["Narendra Modi", "Donald Trump"]


# Step-4 : Load an unknown image to identify the faces

image_to_recognize = face_recognition.load_image_file("trump-modi.jpg")


# Step-5 : Find all the faces and face encodings in the unknown image


# Detect all faces in the image
# Arguments are image, no_of_times_to_upsample, model
all_face_locations = face_recognition.face_locations(image_to_recognize, model="cnn")

# Detect face encodings for all the faces detected
all_face_encodings = face_recognition.face_encodings(image_to_recognize, all_face_locations)

# Print the number of faces detected
print("There are {} face(s) in this image.".format(len(all_face_locations)))


# Step-6 : Loop through face location and encodings

# Loop through each face location and face encodings found in the unknown image

for current_face_location, current_face_encoding in zip(all_face_locations, all_face_encodings):


    # Step-7 : Print locations of each faces (inside the loop)

    # Splitting the tuple to get the four position values
    top_pos, right_pos, bottom_pos, left_pos = current_face_location


    # Step-8 : Compare faces and get the matches list (inside the loop)

    # See if the face is any match(es) for the known face(s)
    all_matches = face_recognition.compare_faces(known_face_encodings, current_face_encoding)


    # Step-9 : Initialize name string (inside the loop)


    # Initialize a name string (inside the loop)
    name_of_person = "Unknown Face"


    # Step-10 : Use first match and get name from the respective index (inside the loop)

    # If a match was found in known_face_encodings, use the first one.
    # Check if all_matches have at least one item
    # If yes, get the index number of face that is located in the first index of all_matches
    if True in all_matches:
        first_match_index = all_matches.index(True)
        name_of_person = known_face_names[first_match_index]


    # Step-11 : Draw rectangle around face (inside the loop)

    # Draw blue rectangle around the face
    cv2.rectangle(original_image, (left_pos, top_pos), (right_pos, bottom_pos), (255, 0, 0), 2)


    # Step-12 : Write name below face (inside the loop)

    font = cv2.FONT_HERSHEY_DUPLEX

    cv2.putText(original_image, name_of_person, (left_pos, bottom_pos), font, 0.5, (255, 255, 255), 1)


    # Step-13 : Show the image with rectangle and text (inside the loop)

    # cv2.imshow('Identified Faces', image_to_recognize)
    cv2_imshow(original_image)
```
